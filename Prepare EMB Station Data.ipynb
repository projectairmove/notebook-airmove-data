```python
all_df = None
```

```python
LOOKUP = {
    'Las Pinas Station': 7,
    'Makati Station': 14,
    'Malabon Station': 13,
    'Mandaluyong Station': 4,
    'Manila Station': 15,
    'Marikina Station': 2,
    'Muntinlupa Station': 3,
    'Navotas Station': 8,
    'North Caloocan Station': 9,
    'Paranaque Station': 10,
    'Pasay Station': 1,
    'Pasig Station': 11,
    'Pateros Station': 6,
    'San Juan Station': 5,
    'Taguig Station': 12,
#     'UE Caloocan Station':
}

for key, value in LOOKUP.items():
    station_df = pd.DataFrame({
        "recorded_time": pd.to_datetime(pm25_df['Date & Time']),
        "pm25": pm25_df[key],
        "pm10": pm10_df[key],
        "station": value,
        "is_published": 1
    }).set_index(['recorded_time', 'station'])
    
#     station_df = station_df[station_df['pm25'].notna() or station_df['pm10'].notna()]
    station_filtered_df = station_df[(station_df['pm25'].notna()) | (station_df['pm10'].notna())]
    station_filtered_df.to_csv(f'data/output/emb/hourly/{key}.csv')
    print(station_filtered_df[:5])
    
    
    if all_df is None:
        all_df = station_filtered_df
    else:
        all_df = pd.concat([all_df, station_filtered_df])
```

```python
all_df.to_csv(f'data/output/emb/hourly/All EMB Stations.csv')
```
