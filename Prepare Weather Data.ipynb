```python
import numpy as np
import pandas as pd
```

### Read stations data

```python
stations_df = pd.read_csv('weather_data/weather_stations.csv')
stations_df
```

### User input data for station_id and filename

```python
station_id = 1
station_name = "NAIA"
filename = station_name + " Daily Data"
print(filename)
```

```python
data_df = pd.read_csv(f'weather_data/{filename}.csv').replace(-999.0, np.nan, regex=True)
data_df
```

```python
data_df.columns = data_df.columns.str.lower()
```

### Creating a Series of datetime objects

```python
datetime_series = pd.to_datetime(data_df[['day', 'month', 'year']].rename(columns={'day': 'day', 'month': 'month', 'year': 'year'}))
```

### Assigning the datetime_series as a new column recorded_time in the DataFrame

```python
data_df['recorded_time'] = datetime_series
data_df['station'] = station_id
data_df = data_df.drop(['year', 'month', 'day'], axis=1)
data_df
```

```python
data_df.to_csv(f'weather_data/output/{filename}.csv')
```
