```python
import numpy as np
import pandas as pd
```

### Import sensor data. Note that you can download this using the airmove dashboard

```python
sensor_data = pd.read_csv('sensor_data/sensor_data_1.csv')
sensor_data
```

### Cleanup data
Check for columns containing only NaN values

```python
# Step 2: Check for columns containing only NaN values
nan_columns = []
for column in sensor_data.columns:
    if sensor_data[column].isnull().all():
        nan_columns.append(column)

# Step 3: Remove columns containing only NaN values
sensor_data = sensor_data.drop(nan_columns, axis=1)

# Step 4: Display the resulting DataFrame
sensor_data
```

### Extract the date component from the datetime column

```python
sensor_data['local_datetime'] = pd.to_datetime(sensor_data['local_time'])
```

### Exclude columns from resampling

```python
# Columns to exclude from resampling
# Set datetime column as index
sensor_data.set_index('local_datetime', inplace=True, drop=False)
extra_columns = sensor_data[['gps_lat', 'gps_long']]
extra_columns
```

```python
sensor_data
```

```python
columns_to_exclude = ['gps_lat', 'gps_long']
sensor_data_dropped = sensor_data.drop(columns=columns_to_exclude)
sensor_data_dropped
```

### Aggregate data using date and get its minimum, maximum, mean, and standard deviation

```python
# Aggregate the data using date and get its min, max, mean, and standard deviation
sensor_data_min = sensor_data_dropped.resample('D', on='local_datetime').min()
sensor_data_max = sensor_data_dropped.resample('D', on='local_datetime').max()
sensor_data_mean = sensor_data_dropped.resample('D', on='local_datetime').mean()
sensor_data_std = sensor_data_dropped.resample('D', on='local_datetime').std()
```

```python
print("sensor_data_min")
sensor_data_min
```

```python
# Drop the following columns as they exist already in sensor_data_min
sensor_data_max = sensor_data_max.drop(columns=["local_time", "source", "local_datetime"])
print("sensor_data_max")
sensor_data_max
```

```python
print("sensor_data_mean")
sensor_data_mean
```

```python
print("sensor_data_std")
sensor_data_std
```

### Combine these data into one array

```python
# Rename columns with "_min", "_max", "_std" suffixes
columns_to_retain = ['local_time', 'source', 'local_datetime']

sensor_data_min.columns = [f"{col}_min" if col not in columns_to_retain else col for col in sensor_data_min.columns]
sensor_data_max.columns = [f"{col}_max" if col not in columns_to_retain else col for col in sensor_data_max.columns]
sensor_data_mean.columns = [f"{col}_std" if col not in columns_to_retain else col for col in sensor_data_mean.columns]
sensor_data_std.columns = [f"{col}_std" if col not in columns_to_retain else col for col in sensor_data_std.columns]
```

```python
# Combine DataFrames into one
daily_data_combined = pd.concat([sensor_data_min, sensor_data_max, sensor_data_mean, sensor_data_std], axis=1)

# Print the combined DataFrame
print(daily_data_combined)
```

```python
# Concatenate extra columns back to the result
daily_data_combined = pd.concat([daily_data_combined, extra_columns.reindex(daily_data_combined.index)], axis=1)
daily_data_combined
```

```python
daily_data_combined.to_csv(f'sensor_data/output/sensor_daily_data.csv')
```
