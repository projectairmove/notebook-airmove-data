## Good enough! Now we automate it.

```python
import datetime as dt
import time

URL_FORMAT = "https://weather.uwyo.edu/cgi-bin/wyowx.fcgi?TYPE=sflist&DATE={date}&HOUR=00&UNITS=M&STATION=RPLL"
# REGEX_FORMAT = r"^(?P<station>.{4}) (?P<month>\d{2})/(?P<hour>\d{2})(?P<min>\d{2}) (?P<altm>.{6}) (?P<tmp>.{3}) (?P<dew>.{3}) (?P<rh>.{3}) (?P<dir>.{3}) (?P<spd>.{3}) (?P<vis>.{4}) (?P<clouds>.{7}).*$"
REGEX_FORMAT_NO_GUS = r"^(?P<station>.{4}) (?P<time>.{7}) (?P<altm>.{6}) (?P<tmp>.{3}) (?P<dew>.{3}) (?P<rh>.{3}) (?P<dir>.{3}) (?P<spd>.{3}).*$"
REGEX_FORMAT_NO_GUS = r"^(?P<station>.{4}) (?P<month>\d{2})/(?P<hour>\d{2})(?P<min>\d{2}) (?P<altm>.{6}) (?P<tmp>.{3}) (?P<dew>.{3}) (?P<rh>.{3}) (?P<dir>.{3}) (?P<spd>.{3}).*$"

start_date = dt.datetime(2017, 1, 1)
end_date = dt.datetime(2022, 7, 14)
delta = dt.timedelta(days=1)

station_data = []

while start_date <= end_date:
    url = URL_FORMAT.format(date=start_date.strftime("%Y%m%d"))
    print(url)
    
    resp = requests.get(url, verify = False)

    with open("data/Climate Data - UWYO/Raw/{}.txt".format(start_date.strftime("%Y%m%d")), "w") as f:
        f.write(resp.text)

    try:
        soup = BeautifulSoup(resp.text)
        soup.findAll("pre")[0]

        clean_soup = soup.findAll("pre")[0].contents[0].split("\n")[4:-1]

        for row in clean_soup:

            row_obj = re.match(REGEX_FORMAT_NO_GUS, row).groupdict()

            # Their date format does not include month
            row_obj['date'] = (start_date + dt.timedelta(hours=int(row_obj['hour']))).strftime("%Y-%m-%dT%H:%M:%S")
            station_data.append(row_obj)
    except:
        print(resp.text)
    
    start_date += delta

#     time.sleep(1) # Don't flood them
```

```python
station_data_df = pd.DataFrame.from_dict(station_data)
station_data_df
```

Data is originally parsed as string. We convert them to integers and floats as necessary.
